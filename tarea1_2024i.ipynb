```python
fpromotores = io.open('promotores.txt','rU')

def cuenta_promotores(sec):
    return len(re.split(r"[AT]GATA[AG]", sec)) -1

for l in fpromotores:
    linea = l.strip()
    llista = re.split(r"\s+", linea)
    print(llista[0]," ", end=": ")
    secuencia = llista[1]
    print(cuenta_promotores(secuencia))
```

```python
from sklearn.metrics import normalized_mutual_info_score
from numpy.random import choice
from matplotlib import pyplot as plt

secuencia = [choice(list('ACGT')) for _ in range(200)]
hebra = list('GATTACA')
N = len(hebra)
```

```python
I = []
for i in range(len(secuencia) - N):
    score = normalized_mutual_info_score(hebra,secuencia[i:i+N])
    I.append(score)
    
fig = plt.figure(figsize=(10,3))
plt.plot(I)
```
